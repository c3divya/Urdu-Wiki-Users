# Social Data Analysis: Urdu User Wiki Edits

```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as py
import numpy as np
import matplotlib.pyplot as plt
```

```python
reader=pd.read_csv("urwiki.csv",low_memory=False)
reader.head()
```

```python
reader_data=reader[['date_time','editor','editor_id','articleid']]
Unique_editors=reader_data['editor'].unique()
```

```python
for x in reader_data.date_time:
    datetime=x.split(' ')
reader_data[['date_time','time']] = reader_data['date_time'].str.split(' ',expand=True)
reader_data.head()
```

# Unique Editors

```python
Unique_editors=pd.DataFrame(reader_data['editor'].unique(),columns=['editor'])
Unique_editors
```

```python
reader_data[['month','date','year']] =reader_data['date_time'].str.split('/',expand=True)
reader_data.head()
```

# creating cohort

```python
Unique_editor_new=reader_data.drop_duplicates(subset='editor')
Unique_editor_new.count
cohart_new=Unique_editor_new.groupby(by=['year'])
cohart_new.head()
```

```python
cohart_2004=reader_data[reader_data.year=='2004']
cohart_2004
```

```python
cohart_dataframe=pd.DataFrame({'Artcle_edited_in_a_month': reader_data.groupby(["year","month","editor"]).articleid.nunique()}).reset_index()
cohart_dataframe
```

```python

cohart_datetimetable=pd.DataFrame({'date_time':reader_data.date_time,'editor':reader_data.editor,'year':reader_data.year,'month':reader_data.month})
cohart_datetimetable
```

```python
result_set=pd.merge(cohart_dataframe,cohart_datetimetable,how='inner',on=['year','editor'])
result_set
```

```python
Cohart_editsinamonth=pd.DataFrame({'Count': cohart_dataframe.groupby(['year','month']).Artcle_edited_in_a_month.sum()}).reset_index()
Cohart_editsinamonth
```

```python
result=pd.merge(cohart_dataframe,Cohart_editsinamonth, how='left', on=['year','month'])
result
```
